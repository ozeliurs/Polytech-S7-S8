# TD5 - Linear and Logistic Regression

```python
from sklearn import datasets, linear_model, model_selection, metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_diabetes, y_diabetes = datasets.load_diabetes(return_X_y=True)

print(f"X_diabetes.shape = {X_diabetes.shape}")
print(f"y_diabetes.shape = {y_diabetes.shape}")

columns = ["Age", "Sex", "Body Mass Index", "Average Blood Pressure", "Total Serum Cholesterol", "Low-density Lipoproteins", "High-density Lipoproteins", "Total Cholesterol", "Log of Serum Triglycerides Level", "Blood Sugar Level"]

for column, field in zip(columns, X_diabetes[0]):
    print(f"{column} = {field}")
print(f"y_diabetes = {y_diabetes[0]}")

# Create subplots for each column
axs = plt.subplots(2, 5, figsize=(20, 10))[1].flatten()

for field, column, ax in zip(X_diabetes.T, columns, axs):
    ax.scatter(field, y_diabetes)
    ax.set_title(column)
```

```python
X_diabetes_train, X_diabetes_test, y_diabetes_train, y_diabetes_test = model_selection.train_test_split(X_diabetes, y_diabetes, test_size=0.3, random_state=0)
```

## Linear Regression

```python
# Draw the points and the trained model
axs = plt.subplots(2, 5, figsize=(20, 10))[1].flatten()

for i in range(len(X_diabetes[0])):
    X_diabetes_train_i = X_diabetes_train[:, i].reshape(-1, 1)
    X_diabetes_test_i = X_diabetes_test[:, i].reshape(-1, 1)

    lr = linear_model.LinearRegression()
    lr.fit(X_diabetes_train_i, y_diabetes_train)

    y_diabetes_pred = lr.predict(X_diabetes_test_i)

    axs[i].scatter(X_diabetes_train_i, y_diabetes_train)
    axs[i].scatter(X_diabetes_test_i, y_diabetes_test)
    axs[i].plot(X_diabetes_test_i, y_diabetes_pred, color='red')
    axs[i].set_title(f"{columns[i]} - {round(lr.score(X_diabetes_test_i, y_diabetes_test), 2)}")
```

```python
X_digit, y_digit = datasets.load_digits(return_X_y=True)

print(f"X_digit.shape = {X_digit.shape}")
print(f"y_digit.shape = {y_digit.shape}")

X_digit_train, X_digit_test, y_digit_train, y_digit_test = model_selection.train_test_split(X_digit, y_digit, test_size=0.3, random_state=0)
```

## Logistic Regression

```python
axs = plt.subplots(10, 10, figsize=(20, 20))[1]

# For each digit pair, train a logistic regression model and plot the confusion matrix
for i in range(10):
    for j in range(i + 1, 10):
        X_digit_train_ij = X_digit_train[np.logical_or(y_digit_train == i, y_digit_train == j)]
        y_digit_train_ij = y_digit_train[np.logical_or(y_digit_train == i, y_digit_train == j)]

        X_digit_test_ij = X_digit_test[np.logical_or(y_digit_test == i, y_digit_test == j)]
        y_digit_test_ij = y_digit_test[np.logical_or(y_digit_test == i, y_digit_test == j)]

        lor = linear_model.LogisticRegression(max_iter=10000)
        lor.fit(X_digit_train_ij, y_digit_train_ij)

        y_digit_pred_ij = lor.predict(X_digit_test_ij)

        print(f"lor.score(X_digit_test_ij, y_digit_test_ij) = {lor.score(X_digit_test_ij, y_digit_test_ij)}")

        mat = metrics.ConfusionMatrixDisplay.from_predictions(y_digit_test_ij, y_digit_pred_ij, display_labels=[i, j]).plot(ax=axs[i, j])
        axs[i, j].set_title(f"{i} vs {j}")
        
plt.show()
```

```python
X_wine, y_wine = datasets.load_wine(return_X_y=True)

print(f"X_wine.shape = {X_wine.shape}")
print(f"y_wine.shape = {y_wine.shape}")

X_wine_train, X_wine_test, y_wine_train, y_wine_test = model_selection.train_test_split(X_wine, y_wine, test_size=0.3, random_state=0)
```

```python
axs = plt.subplots(3, 3, figsize=(20, 20))[1]

# For each digit pair, train a logistic regression model and plot the confusion matrix
for i in range(3):
    for j in range(i + 1, 3):
        X_wine_train_ij = X_wine_train[np.logical_or(y_wine_train == i, y_wine_train == j)]
        y_wine_train_ij = y_wine_train[np.logical_or(y_wine_train == i, y_wine_train == j)]

        X_wine_test_ij = X_wine_test[np.logical_or(y_wine_test == i, y_wine_test == j)]
        y_wine_test_ij = y_wine_test[np.logical_or(y_wine_test == i, y_wine_test == j)]
        
        print(f"X_wine_train_ij.shape = {X_wine_train_ij.shape}")
        print(f"y_wine_train_ij.shape = {y_wine_train_ij.shape}")

        lor = linear_model.LogisticRegression(max_iter=10000)
        lor.fit(X_wine_train_ij, y_wine_train_ij)

        y_wine_pred_ij = lor.predict(X_wine_test_ij)

        print(f"lor.score(X_wine_test_ij, y_wine_test_ij) = {lor.score(X_wine_test_ij, y_wine_test_ij)}")

        mat = metrics.ConfusionMatrixDisplay.from_predictions(y_wine_test_ij, y_wine_pred_ij, display_labels=[i, j]).plot(ax=axs[i, j])
        axs[i, j].set_title(f"{i} vs {j}")
```
